```python
x = [1,2,3,4,5]
y = [2,5,3,8,7]

plt.figure(figsize=(8,5))

plt.scatter(
    x,
    y,
    color="dodgerblue",
    s=120,
    marker="o",
    edgecolors="black",
    linewidths=1.5,
    alpha=0.8,
    label="Students"
)

plt.title("Study Hours vs Marks", fontsize=16)
plt.xlabel("Study Hours", fontsize=12)
plt.ylabel("Marks", fontsize=12)

plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()

for i in range(len(x)):
    plt.text(x[i], y[i]+0.2, f"{y[i]}", ha="center")

plt.tight_layout()
plt.show()
```
